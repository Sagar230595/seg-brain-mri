```python
# importing Zipfile library
# https://stackoverflow.com/questions/3451111/unzipping-files-in-python
import zipfile
with zipfile.ZipFile('/content/kaggle_3m.zip', 'r') as zip_ref:
    zip_ref.extractall('/content')
```

```python
import numpy as np
import pandas as pd
import os
import cv2
import matplotlib.pyplot as plt
from PIL import Image
```

```python
# creating a dataframe which stores path of the images, path of the masks

image_path = []
mask_path = []
for i in sorted(os.listdir('/content/kaggle_3m/')):
  for j in sorted(os.listdir('/content/kaggle_3m/'+i)):
    if 'mask' in j:
      mask_path.append('/content/kaggle_3m/'+i+'/'+j)
    else:
      image_path.append('/content/kaggle_3m/'+i+'/'+j)
```

```python
sorted_masks = {}
for i in range(len(mask_path)):
    sorted_masks[image_path.index(mask_path[i].replace('_mask', ''))] = mask_path[i]
```

```python
masks_list = []

for i in sorted(sorted_masks):
    masks_list.append(sorted_masks[i])
```

```python
path_df = pd.DataFrame(columns=['images', 'masks'])
path_df['images'] = image_path
path_df['masks'] = masks_list
```

```python
# let's make a column in dataframe which has enteries of Yes and No. 'Yes' if patient diagnostic positive and 'No' if patient diagnostic negative

diagnostic_pos_neg = []

for i in range(len(path_df)):
  if np.max(cv2.imread(path_df['masks'][i])) > 0:
    diagnostic_pos_neg.append('Yes')
  else:
    diagnostic_pos_neg.append('No')
```

```python
path_df['diagnostic_pos_neg'] = diagnostic_pos_neg
```

```python
# Let's save this Dataframe.
path_df.to_csv('Final_path_df.csv')
```

```python
# let's create list of path of images and masks.
image_path_list_yes = []
image_path_list_No = []
masks_path_list_yes = []
masks_path_list_No = []

for i in range(len(path_df)):
  if path_df['diagnostic_pos_neg'][i] == 'Yes':
    image_path_list_yes.append(path_df['images'][i])
    masks_path_list_yes.append(path_df['masks'][i])
  else:
    image_path_list_No.append(path_df['images'][i])
    masks_path_list_No.append(path_df['masks'][i])
```

```python
from google.colab.patches import cv2_imshow
```

```python
# let's plot sample positive and negative diagnostic images with it's masks.
# https://www.kaggle.com/code/bonhart/brain-mri-data-visualization-unet-fpn

from mpl_toolkits.axes_grid import ImageGrid

fig = plt.figure(figsize=(25., 25.))
grid = ImageGrid(fig, 111,  # similar to subplot(111)
                 nrows_ncols=(5, 2),  # creates 2x2 grid of axes
                 axes_pad=0.1,  # pad between axes in inch.
                 )
# let's plot positive diagnostic patients brain MRI
for i in range(0,10,2):
  grid[i].imshow(cv2.imread(image_path_list_yes[i], cv2.IMREAD_COLOR))
  grid[i].set_title("Positive", fontsize=15)
  grid[i].axis("off")
  grid[i+1].imshow(cv2.imread(masks_path_list_yes[i], cv2.IMREAD_COLOR))
  grid[i+1].set_title("Positive_masks", fontsize=15)
  grid[i+1].axis("off")
```

In above images, we can clearly see that, In MRI image green spot is exactly matching with mask image of white spot.

```python
fig = plt.figure(figsize=(25., 25.))
grid = ImageGrid(fig, 111,  # similar to subplot(111)
                 nrows_ncols=(5, 2),  # creates 2x2 grid of axes
                 axes_pad=0.1,  # pad between axes in inch.
                 )
# let's plot negative diagnostic patients brain MRI
for i in range(0,10,2):
  grid[i].imshow(cv2.imread(image_path_list_No[i], cv2.IMREAD_COLOR))
  grid[i].set_title("Negative", fontsize=15)
  grid[i].axis("off")
  grid[i+1].imshow(cv2.imread(masks_path_list_No[i], cv2.IMREAD_COLOR))
  grid[i+1].set_title("Negative_masks", fontsize=15)
  grid[i+1].axis("off")
```

In above images we can see that the patient with negative brain MRI images do not have green spot. So, masks image is also fully black. it do not posses white spot in it.
